```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

# Insercion

```python
T_I = pd.read_csv("T_I.csv")
T_I.head()
```

```python
plt.plot(range(1, 501), T_I.mean())
plt.title("Tiempo medio de inserción (ns) para insertar conjuntos\n de 10 objetos en Arbol-R Dinámico de Hilbert")
plt.xlabel("Conjunto C insertado")
plt.ylabel("Tiempo de inserción (ns)")
```

# Borrado

```python
df_eliminacion = pd.read_csv("T_D.csv")
df_eliminacion.head()
```

```python
plt.plot(range(1, 6), df_eliminacion.mean())
plt.title("Tiempo medio de eliminación (ns) para eliminar conjuntos\n de 1000 elementos en Arbol-R Dinámico de Hilbert")
plt.xlabel("Conjunto C eliminado")
plt.ylabel("Tiempo de eliminación (ns)")
```

# KNN

```python
import pandas as pd

df_knn = pd.read_csv("T_K.csv")
df_knn.head()
print(df_knn.shape)
```

```python
df_knn.mean()
```

```python
medias = df_knn.mean()

plt.plot(range(1,6),medias[0:df_knn.shape[1]:3], label="K=1")
plt.plot(range(1,6),medias[1:df_knn.shape[1]:3], label="K=5")
plt.plot(range(1,6),medias[2:df_knn.shape[1]:3], label="K=20")
plt.legend()
plt.title("Tiempo medio de búsqueda KNN (ns) para buscar\n 1, 5, y 20 vecinos por conjuntos insertados \nde 1000 elementos en Arbol-R Dinámico de Hilbert")
plt.xlabel("Conjunto C insertado")
plt.ylabel("Tiempo de búsqueda (ns)")
```
